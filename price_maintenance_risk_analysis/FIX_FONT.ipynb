# 🔧 中文字体修复方案

如果你看到的图表中中文显示为方框乱码，请按以下步骤操作：

## 步骤 1: 运行此单元格初始化字体

````python
import sys
sys.path.append('..')

from utils.font_manager import get_font_prop, init_font

# 初始化字体
font_prop = init_font()

print("\n字体初始化完成！")
print("\n使用方法:")
print("  在绘图时，为每个文本元素添加 fontproperties=font_prop 参数")
print("  例如: ax.set_title('标题', fontproperties=font_prop)")
````

## 步骤 2: 测试字体

````python
import matplotlib.pyplot as plt
import numpy as np

# 测试绘图
fig, ax = plt.subplots(figsize=(10, 6))

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), label='正弦波')
ax.plot(x, np.cos(x), label='余弦波')

# 设置标题和标签（注意：每个文本都需要添加 fontproperties=font_prop）
ax.set_title('中文显示测试 - 应该正常显示', fontproperties=font_prop, fontsize=14)
ax.set_xlabel('横轴标签', fontproperties=font_prop, fontsize=12)
ax.set_ylabel('纵轴标签', fontproperties=font_prop, fontsize=12)

# 设置图例字体
legend = ax.get_legend()
if legend:
    for text in legend.get_texts():
        text.set_fontproperties(font_prop)

ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("✅ 如果上面的图表中文显示正常，说明字体配置成功！")
````

## 💡 提示

如果上面的测试仍然显示乱码，请运行以下命令复制系统字体：

````python
# 复制 macOS 系统字体到项目目录
import shutil
import os

# 创建 fonts 目录
fonts_dir = '../fonts'
os.makedirs(fonts_dir, exist_ok=True)

# macOS 系统字体路径
system_fonts = [
    '/System/Library/Fonts/STHeiti Medium.ttc',
    '/System/Library/Fonts/PingFang.ttc',
]

for src in system_fonts:
    if os.path.exists(src):
        dest = os.path.join(fonts_dir, os.path.basename(src))
        if not os.path.exists(dest):
            shutil.copy(src, dest)
            print(f"✅ 已复制: {os.path.basename(src)}")
        else:
            print(f"⏭️  已存在: {os.path.basename(src)}")
        break

print("\n请重新运行上面的测试单元格（步骤2）")
````
